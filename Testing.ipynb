```python
ticker = 'JY1 Curncy'
testfile = r'G:\Projects\BackTesting1.0\BackTests\ratioMR_rsi_long_TestMR_e785abfb43_438d4e63ac\backtest_results\NQ1 Index_backtest.parquet'
testfile2 = r'G:\Projects\BackTesting1.0\BackTests\ratioMR_rsi_long_TestMR_e785abfb43_438d4e63ac\backtest_results_detail\NQ1 Index_backtest_trades.parquet'
df = pl.read_parquet(testfile2)

df_p = df.to_pandas()

display(df_p)

print(df)

df_p.to_clipboard()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Convert Polars DataFrame to pandas
df_pandas = df.to_pandas()
df_pandas.set_index("Date", inplace=True)

# Create a figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add trace for "Carry" on the primary y-axis
fig.add_trace(
    go.Scatter(x=df_pandas.index, y=df_pandas["Carry"], name="Carry", line=dict(color="blue")),
    secondary_y=False
)

# Add trace for "Close" on the secondary y-axis
fig.add_trace(
    go.Scatter(x=df_pandas.index, y=df_pandas["Close"], name="Close", line=dict(color="red")),
    secondary_y=True
)

# Set titles
fig.update_layout(
    title_text="Carry and Close Price Over Time"
)

# Set x-axis title
fig.update_xaxes(title_text="Date")

# Set y-axes titles
fig.update_yaxes(title_text="Carry", secondary_y=False)
fig.update_yaxes(title_text="Close", secondary_y=True)

# Show plot
fig.show()
```

```python
import plotly.graph_objects as go

# Convert Polars DataFrame to pandas if needed
df_pandas = df.to_pandas()

# Create OHLC chart
fig = go.Figure(data=[go.Ohlc(
    x=df_pandas['Date'],
    open=df_pandas['Open'],
    high=df_pandas['High'],
    low=df_pandas['Low'],
    close=df_pandas['Close'],
    increasing_line_color='green',  # Color for increasing bars
    decreasing_line_color='red'     # Color for decreasing bars
)])

# Set title and labels
fig.update_layout(
    title="OHLC Chart",
    xaxis_title="Date",
    yaxis_title="Price",
)

# Show the plot
fig.show()
```
